# I. Grab the Iris data set and insert into a data frame.

```python
df = pd.read_json('iris.json')
pd.set_option('display.max_rows',151)
df
```

# II. Check shape of data, data types, clean up if necessary.

```python
df.shape
df.dtypes
```

# III. Split the iris species into 3 separate data frames.

```python
df = df.loc[:,['sepalWidth','petalWidth', 'species']].sort_values('petalWidth')

criteria_1 = df['species'] == 'setosa'
criteria_2 = df['species'] == 'versicolor'
criteria_3 = df['species'] == 'virginica'

dfSetosa = df[criteria_1]
dfVersicolor = df[criteria_2]
dfVirginica = df[criteria_3]

dfSetosa.head()
dfVersicolor.head()
dfVirginica.head()
```

# IV. Visualize data and provide statistical analysis.

```python
#TODO: Convert plot to Box and Whisker.

xSetosa = dfSetosa.loc[:,'species']
xVeriscolor = dfVersicolor.loc[:,'species']
xVirginica = dfVirginica.loc[:,'species']

SetosaP = dfSetosa.loc[:,'petalWidth']
VeriscolorP = dfVersicolor.loc[:,'petalWidth']
VirginicaP = dfVirginica.loc[:,'petalWidth']

SetosaS = dfSetosa.loc[:,'sepalWidth']
VeriscolorS = dfVersicolor.loc[:,'sepalWidth']
VirginicaS = dfVirginica.loc[:,'sepalWidth']

fig, (ax1,ax2) = plt.subplots(1,2, figsize=(15,5))

ax1.plot(xSetosa,SetosaP,label = 'Setosa')
ax1.plot(xVeriscolor,VeriscolorP,label = 'Veriscolor')
ax1.plot(xVirginica,VirginicaP,label = 'Virginica')

ax1.set_title('Iris Data: Petal, 1935')
ax1.set_xlabel('Species')
ax1.set_ylabel('Petal Width (cm)')

ax2.plot(xSetosa,SetosaS,label = 'Setosa')
ax2.plot(xVeriscolor,VeriscolorS,label = 'Veriscolor')
ax2.plot(xVirginica,VirginicaS,label = 'Virginica')

ax2.set_title('Iris Data: Sepal, 1935')
ax2.set_xlabel('Species')
ax2.set_ylabel('Sepal Width (cm)')

ax1.legend()
ax2.legend()

plt.tight_layout()
plt.show()

spStat = 'Setosa: Petal\n' + str(SetosaP.describe())
print(spStat)

vepStat = 'Versicolor: Petal\n' + str(VeriscolorP.describe())
print(vepStat)

vipStat = 'Virginica: Petal\n' + str(VirginicaP.describe())
print(vipStat)

ssStat = 'Setosa: Sepal\n' + str(SetosaS.describe())
print(ssStat)

vesStat = 'Versicolor: Sepal\n' + str(VeriscolorS.describe())
print(vesStat)

vipStat = 'Virginica: Sepal\n' + str(VirginicaS.describe())
print(vipStat); #Adding a semicolon at the end of the line cleans up the output


```

```python

setosaSW = dfSetosa.loc[:,'sepalWidth']
setosaPW = dfSetosa.loc[:,'petalWidth']

veriscolorSW = dfVersicolor.loc[:,'sepalWidth']
veriscolorPW = dfVersicolor.loc[:,'petalWidth']

virginicaSW = dfVirginica.loc[:,'sepalWidth']
virginicaPW = dfVirginica.loc[:,'petalWidth']

fig, (ax1,ax2,ax3) = plt.subplots(1,3,figsize=(15,5))

ax1.scatter(setosaSW,setosaPW, color='red')
ax2.scatter(veriscolorSW,veriscolorPW, color='blue')
ax3.scatter(virginicaSW,virginicaPW, color ='purple')

ax1.set_title('Setosa: Sepal/Petal Width')
ax2.set_title('Veriscolor: Sepal/Petal Width')
ax3.set_title('Virginica: Sepal/Petal Width')

ax1.set_xlabel('Sepal Width')
ax2.set_xlabel('Sepal Width')
ax3.set_xlabel('Sepal Width')

ax1.set_ylabel('Petal Width')
ax2.set_ylabel('Petal Width')
ax3.set_ylabel('Petal Width')

plt.tight_layout()
plt.show()

display('Setosa-Sepal/Petal Width Correlation Coefficient: ' + str(np.corrcoef(setosaSW,setosaPW)[1,0]) )
display('Veriscolor-Sepal/Petal Width Correlation Coefficient: ' + str(np.corrcoef(veriscolorSW,veriscolorPW)[1,0]) )
display('Virginica-Sepal/Petal Width Correlation Coefficient: ' + str(np.corrcoef(virginicaSW,virginicaPW)[1,0]) );
```
