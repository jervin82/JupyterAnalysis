```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
%matplotlib inline

plt.style.use('ggplot')
```

```python
#Sources: 
    #https://www.kaggle.com/rtatman/iris-dataset-json-version
    #Anderson, Edgar (1935). The irises of the Gaspe Peninsula, Bulletin of the American Iris Society, 59, 2–5.
#Goal: Grab, clean up, parse JSON data set

df = pd.read_json('iris.json')
pd.set_option('display.max_rows',151)
df
```

```python
#Splitting the different species into separate data frames
df = df.loc[:,['sepalWidth','petalWidth', 'species']].sort_values('petalWidth')

criteria_1 = df['species'] == 'setosa'
criteria_2 = df['species'] == 'versicolor'
criteria_3 = df['species'] == 'virginica'

dfSetosa = df[criteria_1]
dfVersicolor = df[criteria_2]
dfVirginica = df[criteria_3]

dfSetosa.head()
dfVersicolor.head()
dfVirginica.head()
```

```python
#TODO: Convert plot to Box and Whisker.
#      Create function for describe section.

xSetosa = dfSetosa.loc[:,'species']
xVeriscolor = dfVersicolor.loc[:,'species']
xVirginica = dfVirginica.loc[:,'species']

SetosaP = dfSetosa.loc[:,'petalWidth']
VeriscolorP = dfVersicolor.loc[:,'petalWidth']
VirginicaP = dfVirginica.loc[:,'petalWidth']

SetosaS = dfSetosa.loc[:,'sepalWidth']
VeriscolorS = dfVersicolor.loc[:,'sepalWidth']
VirginicaS = dfVirginica.loc[:,'sepalWidth']

fig, (ax1,ax2) = plt.subplots(1,2)

ax1.plot(xSetosa,SetosaP,label = 'Setosa')
ax1.plot(xVeriscolor,VeriscolorP,label = 'Veriscolor')
ax1.plot(xVirginica,VirginicaP,label = 'Virginica')

ax1.set_title('Iris Data: Petal, 1935')
ax1.set_xlabel('Species')
ax1.set_ylabel('Petal Width (cm)')

ax2.plot(xSetosa,SetosaS,label = 'Setosa')
ax2.plot(xVeriscolor,VeriscolorS,label = 'Veriscolor')
ax2.plot(xVirginica,VirginicaS,label = 'Virginica')

ax2.set_title('Iris Data: Sepal, 1935')
ax2.set_xlabel('Species')
ax2.set_ylabel('Sepal Width (cm)')

ax1.legend()
ax2.legend()

plt.tight_layout()
plt.show()

display('Setosa: Petal')
SetosaP.describe()

display('Versicolor: Petal')
VeriscolorP.describe()

display('Virginica: Petal')
VirginicaP.describe()

display('Setosa: Sepal')
SetosaS.describe()

display('Versicolor: Sepal')
VeriscolorS.describe()

display('Virginica: Sepal')
VirginicaS.describe()
```
