```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('fivethirtyeight')
```

```python
#Data gathered from the U.S. Bureau of Labor Statistics
#Transportation by Occupation: Total wage and salary earners 1998 - 2018
pd.read_csv('BLSTransportationSal2.csv')
```

```python
df = pd.read_csv('BLSTransportationSal2.csv')

df.columns = ['series', 'year', 'period', 'label', 'value']
```

```python
df.dtypes
```

```python
df = df[['year','value']]
df
```

```python
x = df.iloc[0:,0]
y = df.iloc[0:,1]
plt.title('Transportation by Occupation: Total Wage/Salary Earners 1998 - 2018')
plt.xlabel('Year')
plt.ylabel('Total Expenditures')
plt.plot(x,y)
plt.gca().set_xticks(df["year"].unique())
plt.tick_params(axis ='x', rotation = 45) 
plt.show()
valOld = df.iloc[0:1,1]
valNew = df.iloc[-1,1]

print('Min:')
print(str(df.iloc[0:,1].min()))

print('Max:')
print(str(df.iloc[0:,1].max()))

print('Percent Change 1998 - 2018:')
pc = ((valNew - valOld)/valOld)*100
print(pc.to_string(index = False)+'%')

```
