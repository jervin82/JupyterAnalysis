### Quick Netflix Tag Analysis Demo

#### This report is simply a quick demonstration of the ability to use Python for analyzing data sets. I will be examining a csv containing the available movies and tv shows on Netflix from 9/27/2021. I will begin by importing the packages necessary for this report and adding the contents of the csv into a dataframe.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('netflix_titles.csv')
df.info()
```

### Data

```python
df
```

### The tags for each entry can be found in the 'listed_in' column:

```python
df.listed_in
```

### For the purposes of this report, I will grab the top ten most frequent tag or tag combinations found in the dataset:

```python
tagCountTopTen = df['listed_in'].value_counts().head(10)
tagCountTopTen.plot(kind='barh', figsize = (15,6))
```
